* 가상화,
  - 가상의 데이터센터를 구축하는 것

  - 실제 물리적 컴퓨터(서버)가 있음

  - personal computer는 접속하는 측 컴퓨터

-- 유연성, 빠르게 대응가능, 스마트폰 상용화 후 접근방법 다양화

고 가용성, 확장성, 재해복구, 유연한 예산 운용, 민첩성, 탄력성, 글로벌 분산, 보안성, 대량의 가용성, 결함에 따른 복구성

### 클라우드 서비스 모델

인터넷을 사용하는 환경

#### 사설 클라우드
사내망, 보안적인 망


#### 하이브리드 클라우드
사내 클라우드를 전용선으로 공용 클라우드와 연결하여 사용하는 환경

Iaas 서비스의 대표적인 가상머신 소프트웨어
하이퍼바이저 : vmware, Hyper-V(윈도우와 통합)

WSL2 : 윈도우에 리눅스 커널이 기본으로 포함
WSL2 기반으로 동작하게 만들어야됨
