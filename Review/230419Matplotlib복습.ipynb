```python
# 데이터 시각화를 위한 도구(MS Power BI, tabler라는 솔루션도 있음)
# 파이썬에서 데이터 시각화를 위해서는 Matplotlib 패키지를 불러와야 한다.

import matplotlib as mpl          # 자체 패키지를 불러오고 mpl로 부르겠다 설정
import matplotlib.pyplot as plt  # 파이썬 특화 matplotlib.pyplot을 불러오고 plt로 부르겠다 설정
```

```python
# Style 지정하는 법
plt.style.use('classic')              # 여러 가지 스타일이 있음, 일종의 테마라고 보면 됨.
import numpy as np

x = np.linspace(0, 10, 100)       # linspace : 원하는 범위의 숫자를 필요한 개수만큼 생성
print(x)
```

```python
#plt.plot(x, x)             # x축 자리에 x, y축 자리에 x를 넣어서 x값이 증가할수록 y값도 같이 증가하는 그래프가 그려진다
print(x)                      # 파형의 진행속도가 다르다=위상 차이가 난다
print(np.sin(x))            # sin x의 값 
```

```python
plt.plot(x, np.sin(x))          # x의 값은 0~10까지 증가하는데 sin x 결과에 따라 y축 값이 아래 그래프와 같이 변한다.
```

```python
plt.plot(x, np.cos(x))          # cos x는 1부터 시작하여 음,양을 왔다갔다 하는 그래프를 생성        
```

```python
# 두 그래프를 하나의 그림판에 동시에 표시하는 법
# 한 셀에 동시에 명령어를 쳐주면 된다.(셀 : 주피터 노트북에서 셀은 코드나 문서, 혹은 둘 다 작성하고 실행하는 공간단위)
# 색상의 디폴트 값은 파란색이다.

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# 그림판을 임의로 지정해서 그리는 방법
fig = plt.figure()                  # fig라는 변수에 빈 그림판을 저장
plt.plot(x, np.sin(x), '-')        # 선의 모양을 다르게 표시해줌. dash(-) 하나는 기본값, 실선(solid line)
plt.plot(x, np.cos(x), '--')      # dash(-) 두개는 dash라고 표현함.
fig.savefig('my_figure.png')  # 그림을 저장하고 싶다면 fig.savefig('원하는 저장위치/파일명.확장자')로 저장할 수 있다.
# 그래프를 표현하고 싶을 때는 png포맷을 가장 많이 사용함. 선의 형태를 명료하게 표현해줌
```

```python
# 한번에 두개의 그래프를 출력하는 방법(하나의 그림판 안에 작은 그림판 두개를 통해 그림판 하나마다 그래프 하나씩 출력하는 법)
plt.figure()                # 그림판을 정의해준다.
```

```python
plt.subplot(2, 1, 1)     # 위치표시(rows(행) , columns(열), 패널의 넘버(몇번째 그림판에 위치시킬건지))
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
# 한 그림판에 여러 그래프 표시, 그래프 색상 지정법
# 빛의 3원색=RGB, 색의 3원색=CMY/K(K가 블랙)
plt.plot(x, np.sin(x), color='green')               # 색상 지정법
plt.plot(x, np.sin(x-1), color='g')                  # 색상 약어로도 지정가능
plt.plot(x, np.sin(x-2), color='0.75')              # 회색조 표시(0~1 사이의 값 지정, 0으로 갈수록 검은색)
plt.plot(x, np.sin(x-3), color='#FACA22')       # 16진수 컬러코드를 이용
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))    #  RGB 수치 조정
plt.plot(x, np.sin(x-5), color='chartreuse')     # 색상의 이름 표현법
```

```python
# linestyle 선 스타일 지정법
plt.plot(x, x+0, linestyle='solid')     # solid 스타일
plt.plot(x, x+1, linestyle='dashed')  # dash 스타일    
plt.plot(x, x+2, linestyle='dashdot') # dashdot 스타일
plt.plot(x, x+3, linestyle='dotted')   # dotted 스타일   

# linestyle 기호를 사용한 스타일 지정법
plt.plot(x, x+4, linestyle='-')           # solid(-) 스타일
plt.plot(x, x+5, linestyle='--')         # dash(--) 스타일 
plt.plot(x, x+6, linestyle='-.')          # dashdot(-.) 스타일
plt.plot(x, x+7, linestyle=':')           # dotted(:) 스타일
```

```python
# 선 스타일과 색상 동시지정법

plt.plot(x, x+0, '-g')      # linestyle solid , color green
plt.plot(x, x+1, '--c')     # linestyle dashed, color cyan
plt.plot(x, x+2, '-.k')     # linestyle dashdot, color black
plt.plot(x, x+3, ':r')      # linestyle dotted, color red
```

```python
# 그림판에서 그래프가 차지하는 범위/사이즈 조정
# 따로 조정을 안하고 실행하면 그림판 사이즈에 빈 공간이 없게끔 보여주려고 함

plt.plot(x, np.sin(x))
plt.xlim(-1, 11)          # x축의 범위를 -1부터 11까지
plt.ylim(-1.5, 1.5)       # y축의 범위를 -1.5부터 1.5까지
# 두 축의 범위를 역순으로 지정하면 각 축방향의 반전을 보여줄 수 있음.
# plt.axis('tight') : 그래프를 그림판에 타이트하게 출력하겠다
# plt.axis('equal') : x, y축의 비율을 동일하게 설정하겠다
# plt.title('A Sine Curve') : 그림판(그래프)의 타이틀명(제목) 지정 
# plt.xlabel('x') : x축에 라벨 표시
# plt.ylabel('sin(x)') : y축에 라벨 표시
# plt.legend() : 범례 표시
```
